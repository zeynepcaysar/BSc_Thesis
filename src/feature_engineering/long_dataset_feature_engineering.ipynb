```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
final_output_path = '/data/caysar9/results/final_long.csv'
```

```python
# put the dataset which comes from long_dataset_preprocess.ipynb file
file_path = '/data/caysar9/results/final_minimum_migraine_sleep.csv'
# Load migraine data
print("Loading data...")
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
# create migraine days per month (the number of unique migraine days)

# creationtime_local is in datetime format
df['starttime_local_to_day'] = pd.to_datetime(df['starttime_local_to_day'])

# Filter to include only rows with a positive `attack_duration_days` to count only actual migraine days
df_with_migraine_days = df[df["attack_duration_days"] > 0]

# Group by `hashed_uid` and `year_month` and count unique migraine days per month
migraine_days_per_month = df_with_migraine_days.groupby(["hashed_uid", "year_month"])["starttime_local_to_day"].nunique().reset_index()

# Rename the column to indicate it represents the count of unique migraine days per month
migraine_days_per_month.rename(columns={"starttime_local_to_day": "migraine_days_per_month"}, inplace=True)

df = df.merge(migraine_days_per_month, on=["hashed_uid", "year_month"], how="left")
```

```python
df.describe()
```

```python
# Convert the 'triggers' column to lowercase
df['triggers'] = df['triggers'].str.lower()

# keyword patterns for each category
# Broad sleep-related words for poor sleep quality
trigger_poor_sleep_keywords = ['sleep', 'nap', 'nightmare', 'restless', 'insomnia', 'couch', 'bed', 'vivid dream','interrupted']

# Broad words for lack of sleep
trigger_lack_sleep_keywords = ['lack of sleep', 'up late', 'late to bed', 'slept late', 'not enough sleep', 'insufficient sleep', 'no sleep', 'woke up early']

# Broad words for physical activity
trigger_physical_activity_keywords = ['physical','activity', 'exertion', 'exercise', 'running', 'walking', 'biking', 'yoga', 'work out', 'muscle']

# Broad words for lack of physical activity
trigger_lack_physical_activity_keywords = ['inactivity', 'no exercise', 'lack of exercise', 'sitting', 'skipped', 'missed work out', 'not enough movement','not enough exercice']

# Broad words for stress-related triggers
trigger_stress_keywords = ['stress', 'anxiety', 'overthinking', 'depression', 'emotional', 'worry', 'overwhelmed', 'nervous', 'mental', 'upset']
```

```python
def create_flag_column(df, column, keywords):
    pattern = '|'.join(keywords)
    return df[column].str.contains(pattern, case=False, na=False).astype(int)

# Create trigger columns
df['trigger_poor_sleep'] = create_flag_column(df, 'triggers', trigger_poor_sleep_keywords)
df['trigger_lack_sleep'] = create_flag_column(df, 'triggers', trigger_lack_sleep_keywords)
df['trigger_physical_activity'] = create_flag_column(df, 'triggers', trigger_physical_activity_keywords)
df['trigger_lack_physical_activity'] = create_flag_column(df, 'triggers', trigger_lack_physical_activity_keywords)
df['trigger_stress'] = create_flag_column(df, 'triggers', trigger_stress_keywords)
```

```python
# Convert the 'reliefs' column to lowercase
df['reliefs'] = df['reliefs'].str.lower()

# keyword patterns for each category
# Broad sleep-related words for sleep 
relief_sleep_keywords = ['sleep', 'nap', 'couch', 'bed','rest','lay']

# Broad words for physical activity
relief_physical_activity_keywords = ['physical','activity', 'exertion', 'exercise', 'running', 'walking', 'walk','biking','bike ride', 'yoga', 'work out', 'muscle','stretch','movement','moving']

# Broad words for lack of physical activity
relief_lack_physical_activity_keywords = ['no movement', 'stop activity', 'stay still', 'no exercice', 'stay motionless', 'minimal activity', 'limited movement']
```

```python
# Create reliefs columns
df['relief_sleep'] = create_flag_column(df, 'reliefs', relief_sleep_keywords)
df['relief_physical_activity'] = create_flag_column(df, 'reliefs', relief_physical_activity_keywords)
df['relief_lack_physical_activity'] = create_flag_column(df, 'reliefs', relief_lack_physical_activity_keywords)
```

```python
df['affected_activities'] = df['affected_activities'].str.lower()
# Affected activities related to Quality of Life
quality_of_life_aff_activities_keywords = ['hard to concentrate','slower [at home]', 'missed work', 'missed family time', 'slower [at work]', 'missed social activity', 'missed school', 'concentration', 'tired', 'unproductive', 'not social', 'no energy', 'no appetite', 'slow/soft [house]', 'slow/soft [work]', 'family time', 'difficulties to concentrate', 'school missed']
```

```python
# Create affected activity QoL column
df['affected_activity_QoL'] = create_flag_column(df, 'affected_activities', quality_of_life_aff_activities_keywords)
```

```python
# Define age bins and labels
bins = [18, 24, 34, 44, 54, 65] 
labels = ["18-24", "25-34", "35-44", "45-54", "55-65"] 

# Create the age_group column
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=True)
```

```python
# Map gender to 1 for Female and 0 for Male, 2 for unknown
df['gender_encoded'] = df['gender'].map({'F': 1, 'M': 0, 'Unknown': 2})
```

```python
# Group by hashed_uid to calculate the mean sleep duration for each user
user_sleep_means = df.groupby('hashed_uid').agg(
    mean_sleep_duration=('sleep_duration_hours', 'mean')
).reset_index()

# Categorize users based on the mean sleep duration thresholds
# Short sleep: less than 7 hours
# Good sleep: between 7 and 9 hours
# Excessive sleep: more than 9 hours
user_sleep_means['sleep_group'] = pd.cut(user_sleep_means['mean_sleep_duration'],
                                         bins=[-np.inf, 7, 9, np.inf],
                                         labels=['Short Sleep', 'Adequate Sleep', 'Excessive Sleep'])

df = pd.merge(df, user_sleep_means[['hashed_uid', 'sleep_group']], on='hashed_uid', how='left')
```

```python
df = df.sort_values(by=['hashed_uid', 'creationtime_local']).reset_index(drop=True)

# Create average Sleep Duration (Past 7 Days)
df['sleep_duration_past_7_days'] = df.groupby('hashed_uid')['sleep_duration_hours'].rolling(window=7, min_periods=1).mean().reset_index(0, drop=True)

# Average Sleep Duration for the Past Month (30 days)
df['sleep_duration_past_month'] = df.groupby('hashed_uid')['sleep_duration_hours'].rolling(window=30, min_periods=1).mean().reset_index(0, drop=True)

# Average Sleep Duration for the Past Semester (180 days)
df['sleep_duration_past_semester'] = df.groupby('hashed_uid')['sleep_duration_hours'].rolling(window=180, min_periods=1).mean().reset_index(0, drop=True)

# Average Sleep Duration for the Past Year (365 days)
df['sleep_duration_past_year'] = df.groupby('hashed_uid')['sleep_duration_hours'].rolling(window=365, min_periods=1).mean().reset_index(0, drop=True)

# Create a count of Past Migraine Attacks (sum of migraine attacks over the past 7 days)
df['migraine_attacks_past7days'] = (df.assign(migraine_occurred=(df['attack_duration_hours'] > 0).astype(int)).groupby('hashed_uid')['migraine_occurred'].rolling(window=7, min_periods=1).sum().reset_index(0, drop=True))

# Create the 'mean_migraine_duration_7days' column (average migraine duration of migraine attacks over the past 7 days)
df['mean_migraine_duration_past7days'] = (df.groupby('hashed_uid')['attack_duration_hours'].rolling(window=7, min_periods=1).mean().reset_index(0, drop=True))                                    
```

```python
# Sort by user and date
df = df.sort_values(by=['hashed_uid', 'creationtime_local']).reset_index(drop=True)

# creationtime_local is in datetime format
df['creationtime_local'] = pd.to_datetime(df['creationtime_local'])

# Define a 7-day time window
time_window = 7

# Function to check for a migraine occurrence within the next 'time_window' days, in my case: 7
def next_migraine_within_window(df, time_window):
    for idx in range(len(df)):
        current_day = df.iloc[idx]['creationtime_local']
        future_rows = df[(df['creationtime_local'] > current_day) & 
                         (df['creationtime_local'] <= current_day + pd.Timedelta(days=time_window))]
        # If any of the future rows have 'attack_duration_hours' > 0, set 'next_migraine_in_window' to 1
        if any(future_rows['attack_duration_hours'] > 0):
            df.iloc[idx, df.columns.get_loc('next_migraine_in_window')] = 1  
        else:
            df.iloc[idx, df.columns.get_loc('next_migraine_in_window')] = 0
    return df

df['next_migraine_in_window'] = 0

# Apply the function for each user
df = df.groupby('hashed_uid').apply(next_migraine_within_window, time_window=time_window)

df.reset_index(drop=True, inplace=True)
```

```python
# Defined thresholds for severe migraine
severe_pain_threshold = 7
severe_duration_threshold = 24  # in hours

# Create a new column 'severe_migraine' based on the thresholds
df['severe_migraine'] = (
    (df['pain_intensity'] >= severe_pain_threshold) &
    (df['attack_duration_hours'] >= severe_duration_threshold)
).astype(int)
```
